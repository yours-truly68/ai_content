```python
import os
from dotenv import load_dotenv

load_dotenv()#loads the environment variables into the MacOS Ram, and in the os.environ terminal space

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent
from langchain_groq import ChatGroq

cloud_llm = ChatGroq(
    model = "llama-3.3-70b-versatile",
    temperature=0
)


def get_weather(city: str) -> str:
    '''Get weather of a city'''
    return f"The weather of {city} is sunny!"


agent = create_agent(
    model = cloud_llm,
    system_prompt = "You are a helpful assistant.",
    tools = [get_weather]
)

agent

```

```python
response = agent.invoke({
    "messages" : [
        {
            "role" : "user",
            "content" : "What is the weather like in Bengaluru?"
        }
    ]
})
```

```python
response["messages"][-2].content # tool response --> 'The weather of Bengaluru is sunny!'
response["messages"][-1].content # AI response --> "I'm glad I could help you with the weather in Bengaluru. If you need any more information or have other questions, feel free to ask!"
```
