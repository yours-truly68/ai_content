### Messages

Messages are the fundamental unit of context for models in LangChain. They represent both content and metadata needed to represent the state of the conversation when interacting with LLMs. Messages are objects that contain 
-> Role - Identifies the message type 
-> Content = represent the actual content of the message
-> Metadata = Optional fields such as message IDs, token usage, response type.

```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model = "groq:qwen/qwen3-32b"
)
response = model.invoke("Explain AI to me like I'm 5")
response
```

```python
from langchain.messages import SystemMessage, AIMessage, ToolMessage, HumanMessage

messages = [
    SystemMessage("You are a poetry expert."),
    HumanMessage("Write a poem on AI")
]

response = model.invoke(messages)
response.content
```

### Detailed information to the LLM through System Message!

```python
from langchain.messages import HumanMessage, SystemMessage

system_prompt = SystemMessage("""You are a senior Python Developer with expertise in web frameworks.
                              Always provide code examples and explain your reasoning.
                              Be concise but thorough in your explanations.
                              """)

messages = [
    system_prompt,
    HumanMessage("Do I learn LangGraph next or RAG?")
]
response = model.invoke(messages)
print(response.content)
```

```python
from ast import List
from pydantic import BaseModel, Field


class Actor(BaseModel):
    """Details of the actor"""
    name: str = Field(description="Name of the actor")
    role: str = Field(description="Role of the actor")

class Movies(BaseModel):
    """Details of the a movie"""
    title: str = Field(description="This is the title of the movie")
    descripton: str = Field(description="Description of the movie")
    director: str = Field(description="This is the director of the movie")
    actors:  list[Actor]
    genre: list[str]
    rating: float = Field(description="This is the movie's rating as per IMDB")
    budget: int = Field(description="Budget of the movie in millions")
    gross_collection: int = Field(description="Total collection of the movie in millions")
    

model_with_structure = model.with_structured_output(Movies)

response = model_with_structure.invoke("Please give me the details of the movie Avengers Endgame")
response
    
    
```
