### Model Integration


```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model#library used to initialise any kind of chat model

# model= init_chat_model("gpt-4.1")

model = init_chat_model(
    model = "groq:llama-3.3-70b-versatile", # best practice, you can also use init_chat_model(model = "llama-3.3-70b-versatile", model_provider = "groq")
)
model
```

### Streaming

streaming is used to showcase the output in chunks instead of waiting for the entire response, making user experience progressively better


```python
question = "Why is the sky blue? but not the sun even tho its the same source why isnt it blue like the rest of the sky?"

model.stream(question)

#.stream() creates iterators you have to yield them 
for chunk in model.stream(question):
    print(chunk.text, end = "", flush=True)
```

### Batching

Batching a collection of independent requests to the model, can significantly improve the performance and reduce costs, as the processing can be done in parallel.


```python
responses = model.batch(
    [
        "The best dessert in the world?",
    "The best restaurant in the world?",
    "The best chef in the world?",
    "Which restaurant has the most michellin stars?",
    "Who is the best chef in the world?"
    
    ],
    config = {
        "max_concurrency" : 5 # maximum number of parallel processing or requests to handle at once
    }
)

for response in responses:
    print(response.content)
```
